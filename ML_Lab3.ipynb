```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("wine.csv") 
```

```python
print(data)
```

```python
scores = pd.DataFrame(data)
scores
```

```python
#column-wise mean
scores.mean(axis=0)
```

```python
#row-wise mean
scores.mean(axis=1)
```

```python
#overall mean 
def mean_overall(scores):
    arr = np.array(scores)
    mn = np.mean(arr)
    print(mn)
mean_overall(scores)
```

```python
#column-wise median
scores.median()
```

```python
#row-wise median
scores.median(axis = 1)
```

```python
#overall median
np.median(scores.values)
```

```python
#row wise mode
scores.mode(axis = 1)
```

```python
#column-wise mode
scores.mode()
```

```python
#row-wise standard deviation
scores.std(axis = 1)
```

```python
#column-wise standard deviation
scores.std(axis = 0)
```

```python
#overall std 
np.std(scores.values)
```
